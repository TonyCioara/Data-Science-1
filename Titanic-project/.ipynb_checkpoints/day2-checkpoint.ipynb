```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv('titanic.csv')
```

```python
titanic_df.head()
```

```python
cabin_series = titanic_df['Cabin']
```

```python
cabin_series.head()
```

```python
deck_list = list()
for cabin in cabin_series:
    if type(cabin) is str:
        deck_list.append(cabin[0])
    else:
        deck_list.append(np.nan)
```

```python
deck_list
```

```python
titanic_df['Deck'] = deck_list
```

```python
titanic_df.head()
```

```python
titanic_df[(titanic_df['Deck'] == 'A') & (titanic_df['Survived']) == 1].head()
```

```python
titanic_df[(titanic_df['Deck'] == 'A') | (titanic_df['Deck'] == 'B')].head()
```

```python
titanic_df.head()
```

```python
fc_people = titanic_df[titanic_df['Pclass'] == 1]
```

```python
fc_survived = fc_people[fc_people['Survived'] == 1]
```

```python
fc_percentage = float(len(fc_survived)) / float(len(fc_people)) * 100
```

```python
fc_percentage
```

```python
total_survived = titanic_df[titanic_df['Survived'] == 1]
```

```python
total_percentage = float(len(total_survived)) / float(len(titanic_df)) * 100
```

```python
total_percentage
```

```python
nofc_people = titanic_df[titanic_df['Pclass'] != 1]
```
